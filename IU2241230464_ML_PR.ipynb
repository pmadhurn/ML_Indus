````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# 1 : Creating data frame
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("First 5 rows of the dataset:")
df.head()
#2 : Reading files
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nData Types:")
print(df.dtypes)
print("\nDataset Info:")
df.info()
````

````python
#3 : Slicing manipulations
# Row slicing
print("Rows 5 to 10:")
df[5:11]
# Column selection
print("Select specific columns:")
df[['Name', 'Age', 'Sex', 'Survived']].head()
````

````python
#4 : Columns and row manipulations with loops
# Find's all passengers with specific criteria
print("Passengers who survived and were in 1st class:")
for index, row in df.iterrows():
    if row['Survived'] == 1 and row['Pclass'] == 1:
        print(f"{row['Name']} - Age: {row['Age']}")
    if index > 10:
        break
````

````python
import numpy as np
import pandas as pd
path = "/content/data.csv"
df = pd.read_csv(path)
print(df)

def find_s(data):
  attributes = list(data.columns[:-1])
  atrr = len(attributes)
  print(f"Number of attributes: {atrr}")
  print(f"Attributes: {attributes}")
  hypothesis = ['0'] * atrr
  print(f"Initial hypothesis h0: {hypothesis}")

  for i, row in data.iterrows():
    if row['Output'] == 'Yes':
      print(f"\nPositive example {i+1}:")
      example = []
      for attr in attributes:
        example.append(row[attr])
      print(f"Values: {example}")

      for j in range(atrr):
        if hypothesis[j] == '0':
          hypothesis[j] = row[attributes[j]]
        elif hypothesis[j] != row[attributes[j]]:
          hypothesis[j] = '?'

      print(f"Updated hypothesis: {hypothesis}")

  return hypothesis

final_hypothesis = find_s(df)
print(f"\n\nFinal hypothesis: {final_hypothesis}")

print("\nFinal hypothesis with attribute names:")
attributes = list(df.columns[:-1])
for i in range(len(attributes)):
  print(f"{attributes[i]}: {final_hypothesis[i]}")
````

Candidate Elimination Algorithm

````python
import numpy as np
import pandas as pd
def candidate_elimination(data):
    attributes = list(data.columns[:-1])
    n_attributes = len(attributes)
    print(f"Attributes: {attributes}\n")
    S = ['0'] * n_attributes
    print(f"Initial S: {S}")
    G = [['?'] * n_attributes]
    print(f"Initial G: {G}\n")
    for index, row in data.iterrows():
        print(f"\nExample {index + 1}: ", end="")
        example = []
        for attr in attributes:
            example.append(row[attr])
        print(f"{example}, Output: {row['Output']}")
        if row['Output'] == 'Yes':
            print("Positive example - Update S and G:")
            for i in range(n_attributes):
                if S[i] == '0':
                    S[i] = row[attributes[i]]
                elif S[i] != row[attributes[i]]:
                    S[i] = '?'
            G_new = []
            for g in G:
                consistent = True
                for i in range(n_attributes):
                    if g[i] != '?' and g[i] != row[attributes[i]]:
                        consistent = False
                        break
                if consistent:
                    G_new.append(g)
            G = G_new
        else:
            print("Negative example - Update G:")
            covers_negative = True
            for i in range(n_attributes):
                if S[i] != '?' and S[i] != row[attributes[i]]:
                    covers_negative = False
                    break
            if covers_negative:
                print("S covers negative example - Algorithm fails!")
                return None, None
            G_new = []
            for g in G:
                covers = True
                for i in range(n_attributes):
                    if g[i] != '?' and g[i] != row[attributes[i]]:
                        covers = False
                        break
                if covers:
                    for i in range(n_attributes):
                        if g[i] == '?':
                            if S[i] != '?':
                                g_specialized = g.copy()
                                g_specialized[i] = S[i]
                                if g_specialized[i] != row[attributes[i]]:
                                    G_new.append(g_specialized)
                else:
                    G_new.append(g)
            G = []
            for g in G_new:
                if g not in G:
                    G.append(g)
        print(f"S: {S}")
        print(f"G: {G}")
    return S, G
print("CANDIDATE ELIMINATION ALGORITHM")
print("="*50)
S_final, G_final = candidate_elimination(df)
print("\n" + "="*50)
print("FINAL RESULT:")
print(f"Final S (Specific boundary): {S_final}")
print(f"Final G (General boundary): {G_final}")
attributes = list(df.columns[:-1])
print("\nFinal hypothesis boundaries:")
print("\nS boundary:")
for i in range(len(attributes)):
    print(f"  {attributes[i]}: {S_final[i]}")
print("\nG boundary:")
for j, g in enumerate(G_final):
    print(f"  Hypothesis {j+1}:")
    for i in range(len(attributes)):
        print(f"    {attributes[i]}: {g[i]}")
````
